```python
from datetime import datetime
import pandas as pd
import time
import re
import os
import openpyxl

start=datetime.strptime("201601","%Y%m")
end=datetime.strptime("202010","%Y%m")
dates=[date.strftime("%Y%m") for date in pd.date_range(start=start,end=end,freq="M",closed="right")]
vege = ['배추', '무', '미나리', '갓', '쪽파','생강']


url="D:/projects/data/235801_2021 농산물 가격예측 AI 경진대회/public_data/train_AT_TSALET_ALL/AT_TSALET_ALL_"
data_list=[]
for date in dates:
    file_name=url+date+'.csv'
    print(file_name)
    temp_data=pd.read_csv(file_name)
    temp_data=temp_data[temp_data['PUM_NM'].isin(vege)] 
    try:
        data_list.append(temp_data)
    except FileNotFoundError as e:
        #국제수산처럼, 파일이 없는 경우.
        continue
    del temp_data
file_name='../files/도매시장/'+'도매시장_2016_2020'+'.csv'
res=pd.concat(data_list)
del data_list
res.to_csv(file_name,index=False)
```
