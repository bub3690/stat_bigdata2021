```python
#칼럼 drop
#총 15, 버리기: 8컬럼
#7컬럼 남아야한다.


#등급 버리기
minari.drop(['LV_NM'], axis=1, inplace=True)
#회사 버리기
minari.drop(['CMP_NM'], axis=1, inplace=True)
#cost 버리기
minari.drop(['COST'], axis=1, inplace=True)
#POJ_NM 묶음 버리기
minari.drop(['POJ_NM'], axis=1, inplace=True)
#KIND_NM 품종 버려도 되나?
minari.drop(['KIND_NM'], axis=1, inplace=True)
#SIZE_NM
minari.drop(['SIZE_NM'], axis=1, inplace=True)
minari.drop(['DANQ'], axis=1, inplace=True)
minari.drop(['QTY'], axis=1, inplace=True)
```

```python
minari.columns
```

```python
#날짜 변경
minari['SALEDATE'] = pd.to_datetime(minari['SALEDATE'], format='%Y%m%d')
#일자 소팅
minari = minari.sort_values(by='SALEDATE')
minari.head()
```

```python
#도수분포?
print(minari['TOT_AMT'].value_counts())
print('음수: ',sum(minari['TOT_AMT']<0))
print('0값 : ',sum(minari['TOT_AMT']==0))# 0인 값도 존재. 평균에 영향
print("TOT_AMT의 NA 수 : ",sum(minari["TOT_AMT"].isna()))
```

```python
filter=minari['TOT_AMT']==0
minari[filter].head() #이상치?
```

```python
filter=minari['TOT_QTY']==0
minari[filter].tail() #이상치?
```

```python
#총금액 음수 및 0원 row 제거
data_filter = (minari['TOT_AMT'] > 0)
minari = minari.loc[data_filter]
```

```python
#총 물량 음수 및 0원 row 제거
data_filter = (minari['TOT_QTY'] > 0)
minari = minari.loc[data_filter]
```

```python
print("TOT_QTY의 NA 수",sum(minari["TOT_QTY"].isna()))
#미나리는 NA가 없다.
```

```python
print("TOT_AMT의 NA 수",sum(minari["TOT_AMT"].isna()))
#미나리는 NA가 없다.
```

```python
# 총 금액/ 총 물량  = 단위 금액
#컬럼 추가
minari.insert(7,"PER_PRICE",minari["TOT_AMT"]/minari["TOT_QTY"])
```

```python
minari
```

```python
#산지 전처리
minari['SAN_NM'] = minari['SAN_NM'].str.split(" ").str[0]
minari['SAN_NM'].unique()     # 중국, nan, - : 제거 
# 울산 -> 울산광역시
# 광주 -> 광주광역시
# 부산 -> 부산광역시
# 서울 -> 서울특별시
# 충남 -> 충청남도
# 경기 -> 경기도
# 전북 -> 전라북도
# 전남 -> 전라남도
# 경남 -> 경상남도
# 경북 -> 경상북도
# 충북 -> 충청북도
# 대전 -> 대전광역시
# 대구 -> 대구광역시
# 강원 -> 강원도
# 제주 -> 제주도
# 중국 -> na
# 인천 -> 인천광역시
```

```python
minari['SAN_NM'].replace(['중국','울산','광주','부산','서울','충남','경기','전북','전남','경남','경북','충북','대전','대구','강원',
                          '제주','인천'],
                      [np.nan, '울산광역시', '광주광역시', '부산광역시','서울특별시', 
                       '충청남도', '경기도','전라북도','전라남도','경상남도','경상북도','충청북도','대전광역시','대구광역시','강원도',
                       '제주도','인천광역시'],
                      inplace=True)
minari = minari.dropna(subset=['SAN_NM'])
print(minari['SAN_NM'].unique())
print(len(minari['SAN_NM'].unique()))
```

```python
minari_mean = pd.DataFrame(minari.iloc[:, 7].groupby(minari['SAN_NM']).agg('mean'))
minari_mean['PER_PRICE'] = round(minari_mean['PER_PRICE'], 0)
minari_mean
```

```python
#컬럼명 변경
minari.columns=["품목","총금액","산지","경락 일자","단위","도매시장","총물량","PER_PRICE"]
minari.columns
```
