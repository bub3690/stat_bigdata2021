```python
from datetime import datetime
import requests
import pandas as pd
url = 'https://www.nongnet.or.kr/api/whlslDstrQr.do?sdate='
def date_range(start, end):
    start = datetime.strptime(start, "%Y%m%d")
    end = datetime.strptime(end, "%Y%m%d")
    dates = [date.strftime("%Y%m%d") for date in pd.date_range(start, periods=(end-start).days+1)]
    return dates

date = date_range("20200929", "20210930")   # 시작날짜, 끝 날짜 사이의 날짜 리스트 생성
vege = ['배추', '무', '미나리', '갓', '쪽파','생강']
data_list=[]

def date_range(start,end):
    start = datetime.strptime(start,"%Y%m%d")
    end = datetime.strptime(end,"%Y%m%d")
    dates = [ date.strftime("%Y%m%d") for date in pd.date_range(start,periods=(end-start).days+1)]
    return dates


for i in date:
    print(i)
    response=requests.get(url+i)
    df=pd.DataFrame(response.json()['data'])
    try:
        df=df[df['PUM_NM'].isin(vege)]
    except:
        print("no data ",i)
    data_list.append(df)

data_concat = pd.concat(data_list, ignore_index=True) 
del data_list
data_concat.to_csv('../files/도매시장/more_data1.csv',index=False)
```
