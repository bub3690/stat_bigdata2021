```python
# webdriver 구동하기
from selenium import webdriver
#샐래니움 패키지에서 webdriver class를 import한다.
import time
import pandas as pd

# 3-2. webdriver path지정
driver=webdriver.Chrome('../driver/chromedriver.exe') #webdriver class에서 chrome함수를 불러, 새로운 객체를 또 만들어준다.
# ./ 는 현재 코드가 실행된 현재 위치. 상대경로라고 한다.
```

```python
#url 제어하기

#네이버 데이터 인사이트로 이동해보자
url="https://datalab.naver.com/shoppingInsight/sCategory.naver"
driver.get(url)
```

```python
#날짜 선택 함수들.

def from_date_click(param_year,param_month,param_day):
    try:
        from_year_box=driver.find_element_by_css_selector(".space_top > .insite_inquiry span:nth-child(1) .w2 .select_btn")
        from_year_box.click()

        #2017년 글자 있는지 탐색
        from_year_box_list= driver.find_element_by_css_selector(".space_top > .insite_inquiry .insite_form .set_period_target span:nth-child(1) div:nth-child(1) .scroll_cst")
        from_year=from_year_box_list.find_element_by_xpath(".//*[text() = '{}']".format(param_year)) #여기가 2017년 선택부분.
        from_year.click()# 2017년 클릭.


        #~월 박스 선택
        from_month_box = driver.find_element_by_css_selector(".space_top > .insite_inquiry span:nth-child(1) .w2+ .w3 .select_btn")
        from_month_box.click()

        from_month_box_list= driver.find_element_by_css_selector(" .insite_inquiry .insite_form .set_period_target span:nth-child(1) div:nth-child(2) .scroll_cst")
        if param_month < 10 :
            from_month =from_month_box_list.find_element_by_xpath(".//*[text() = '{}']".format('0'+str(param_month))) #여기가 2017년 선택부분.
        else:
            from_month =from_month_box_list.find_element_by_xpath(".//*[text() = '{}']".format(param_month)) #여기가 2017년 선택부분.
        from_month.click()

        from_day_box = driver.find_element_by_css_selector(".space_top > .insite_inquiry span:nth-child(1) .w3+ .w3 .select_btn")
        from_day_box.click()

        #1일 부터 31일까지 탐색. (각 달마다 일수가 달라서, 예외처리가 필요함.)
        #day = 1
        from_day_box_list= driver.find_element_by_css_selector(".insite_inquiry .insite_form .set_period_target span:nth-child(1) div:nth-child(3) .scroll_cst")
        if param_day < 10 :
            from_day =from_day_box_list.find_element_by_xpath(".//*[text() = '{}']".format('0'+str(param_day))) #여기가 2017년 선택부분.
        else:
            from_day =from_day_box_list.find_element_by_xpath(".//*[text() = '{}']".format(param_day)) #여기가 2017년 선택부분.
        from_day.click()
    except:
        #윤달, 31없는 경우 등, 예외처리
        print("예외처리")
        return False
    
    return True



def to_date_click(param_year,param_month,param_day):
    try:
        to_year_box=driver.find_element_by_css_selector(".space_top > .insite_inquiry .bar+ span .w2 .select_btn")
        to_year_box.click()

        #2017년 글자 있는지 탐색
        to_year_box_list= driver.find_element_by_css_selector(" .space_top > .insite_inquiry .insite_form .set_period_target .bar+ span div:nth-child(1) .scroll_cst")
        to_year=to_year_box_list.find_element_by_xpath(".//*[text() = '{}']".format(param_year)) #여기가 2017년 선택부분.
        to_year.click()# 2017년 클릭.

        #~월 박스 선택
        to_month_box = driver.find_element_by_css_selector(".space_top > .insite_inquiry .bar+ span .w2+ .w3 .select_btn")
        to_month_box.click()

        to_month_box_list= driver.find_element_by_css_selector(".space_top > .insite_inquiry .insite_form .set_period_target .bar+ span div:nth-child(2) .scroll_cst")
        if param_month < 10 :
            to_month =to_month_box_list.find_element_by_xpath(".//*[text() = '{}']".format('0'+str(param_month))) #여기가 2017년 선택부분.
        else:
            to_month =to_month_box_list.find_element_by_xpath(".//*[text() = '{}']".format(param_month)) #여기가 2017년 선택부분.
        to_month.click()

        to_day_box = driver.find_element_by_css_selector(".space_top > .insite_inquiry .bar+ span .w3+ .w3 .select_btn")
        to_day_box.click()

        #1일 부터 31일까지 탐색. (각 달마다 일수가 달라서, 예외처리가 필요함.)
        #day = 1
        to_day_box_list= driver.find_element_by_css_selector(".space_top > .insite_inquiry .insite_form .set_period_target .bar+ span div:nth-child(3) .scroll_cst")
        if param_day < 10 :
            to_day =to_day_box_list.find_element_by_xpath(".//*[text() = '{}']".format('0'+str(param_day))) #여기가 2017년 선택부분.
        else:
            to_day =to_day_box_list.find_element_by_xpath(".//*[text() = '{}']".format(param_day)) #여기가 2017년 선택부분.
        to_day.click()
    except:
        #윤달, 31없는 경우 등, 예외처리
        print("예외처리")
        return False
    
    return True
    
```

```python
#메인 로직
#09-23. 2017년 데이터 출력하도록 조절.


first = True
#처음 시작할때는 from 부터 시작해야해서 만든 변수.
list_search=[]#검색어를 담는 리스트
list_date=[]#날짜를 담는 리스트


for year in range(2017,2018):
    for month in range(8,13):
        for day in range(1,32):
            #1. 분야 박스 선택.
            category_box=driver.find_element_by_css_selector(".space_top > .insite_inquiry .category .select:nth-child(1) .select_btn")
            category_box.click()

            #2. 패션의류 선택
            category_box_inside=driver.find_element_by_css_selector(" .space_top > .insite_inquiry .category  div:nth-child(1) .scroll_cst li:nth-child(1) .option")
            category_box_inside.click()

            #3. 시작은 ~부터 선택.
            if first:
                # 2017년 부터.
                from_res=from_date_click(year,month,day)
                #4. ~ 까지 선택.
                to_res=to_date_click(year,month,day)
                first=False
            else:
                # 2017년 ~까지 선택.
                from_res=to_date_click(year,month,day)
                to_res=from_date_click(year,month,day)
            if (not from_res) or  (not to_res):
                #윤달또는 31일이 없는경우 등
                print(year,'.',month,'.',day," 라는 날짜는 없습니다")
                continue
            
            #5. 조회하기
            #조회 버튼
            submit_btn=driver.find_element_by_css_selector(".space_top > .insite_inquiry .btn_submit")
            submit_btn.click()

            time.sleep(1)

            #6. 데이터 20개 가져오기.
            data_list=driver.find_elements_by_css_selector(".link_text")

            #selenium객체를 text 리스트로 변환
            data=[x.text for x in data_list]
            data_length=len(data)
            
            #7. 리스트에 데이터 합치기
            date_temp=str(year)+'.'+str(month)+'.'+str(day)
            date=[date_temp,]*data_length
            list_search=list_search+data#[1,2,3,4]+[5,6]=[1,2,3,4,5,6]   
            list_date=list_date+date
            print(date_temp)
            #8. 다음일로 넘어가기.

#8. data concat

DATA=pd.concat([
        pd.DataFrame(list_search),
        pd.DataFrame(list_date),],axis=1)

#6. 컬럼명 변경 , to_excel
DATA.columns= ['search','date']
DATA.to_excel("네이버_검색데이터_2017_패션의류.xlsx")

```
