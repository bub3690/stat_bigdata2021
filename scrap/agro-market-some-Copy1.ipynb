````python
from selenium import webdriver
#샐래니움 패키지에서 webdriver class를 import한다.
import time
import pandas as pd
import re
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import os
import openpyxl
````

````python
# webdriver 구동하기

# webdriver path지정
driver=webdriver.Chrome('../driver/chromedriver.exe') #webdriver class에서 chrome함수를 불러, 새로운 객체를 또 만들어준다.
# ./ 는 현재 코드가 실행된 현재 위치. 상대경로라고 한다.
driver.set_window_size(1024, 600)
````

````python
#url 제어하기

#도매유통 정보시스템으로 이동하자
url="https://at.agromarket.kr/users/price/retail/period.do"
driver.get(url)
````

````python
date_form=['2019/','2020/','2021/']
#년도를 붙어주는 함수.
def make_date(date_idx,date):
    return date_form[date_idx//4]+date
````

````python
## 생강부터 시작.
#메인 로직


list_area=[
"KAMIE_2701",#"세종"
"KAMIE_1101",#"서울"
"KAMIE_2100",#"부산"
"KAMIE_2200",#"대구"
"KAMIE_2300",#"인천"
"KAMIE_2401",#"광주"
"KAMIE_2501",#"대전"
"KAMIE_2601",#"울산"
"KAMIE_3111",#"수원"
"KAMIE_3211",#"춘천"
"KAMIE_3311",#"청주"
"KAMIE_3511",#"전주"
"KAMIE_3711",#"포항"
"KAMIE_3911",#"제주"
"KAMIE_3113",#"의정부"
"KAMIE_3613",#"순천"
"KAMIE_3714",#"안동"
"KAMIE_3814",#"창원"
"KAMIE_3214",#"강릉"
    ]#지역 담는 리스트
list_area_name=[ "세종","서울","부산","대구","인천","광주","대전","울산","수원","춘천",
                "청주","전주","포항","제주","의정부","순천","안동","창원","강릉", ]#지역 이름담는 리스트
list_date = ["startDay=2019-01-01&endDay=2019-04-10",#19년 붙이기
             "startDay=2019-04-11&endDay=2019-07-19",#19
             "startDay=2019-07-20&endDay=2019-10-27",#19
             "startDay=2019-10-28&endDay=2019-12-31",#19
             "startDay=2020-01-01&endDay=2020-04-09",#20
             "startDay=2020-04-10&endDay=2020-07-17",#20
             "startDay=2020-07-18&endDay=2020-10-25",#20
             "startDay=2020-10-26&endDay=2020-12-31",#20
             "startDay=2021-01-01&endDay=2021-04-10",#21
             "startDay=2021-04-11&endDay=2021-07-19",#21
             "startDay=2021-07-20&endDay=2021-10-03"]#21

list_item=["KAMIS_247",]
list_item_name=["생강",]#품목이름을 담는 리스트

list_kind=[["KAMIS_247_00"],]#품종을 담는 리스트

list_kind_name=[
           ["생강"],]#품종의 이름을 담는 리스트


count = 0
for ind_item, item in enumerate(list_item):
    #아이템별 리스트. 하나당 1일씩
    item_name=[]
    item_kind=[]
    item_date=[]
    item_area=[]
    item_mean=[]#해당날짜 평균값들
    
    for ind_kind,kind in enumerate(list_kind[ind_item]):
        for ind_area,area in enumerate(list_area):
            for ind_date,date in enumerate(list_date):
                #print(list_item_name[ind_item],kind,area,date)
                #1.url 조합. 이동
                url1="https://at.agromarket.kr/users/price/retail/period.do?productClsCode=01&"
                url2=date#  날짜
                url3="&countryCode="+area # + 지역
                url4="&itemCategoryCode=KAMIS_200"# + 채소
                url5="&itemCode="+item
                url6="&kindCode="+kind# 품종
                url1=url1+url2+url3+url4+url5+url6+"&productRankCode=KAMIS_PRRANK_04&convertKgYn=Y"
                driver.get(url1)
                try:
                    WebDriverWait(driver,timeout=5).until(EC.presence_of_element_located((By.CSS_SELECTOR, "th+ th")))
                except:
                    #데이터가 없는 경우는. 다음으로.
                    continue
                #데이터 수집하기.
                
                #날짜들
                temp_date=driver.find_elements_by_css_selector("th+ th")
                temp_len=len(temp_date)
                item_date+=[make_date(ind_date,x.text) for x in temp_date]
                
                #평균값들
                item_mean+=[x.text for x in driver.find_elements_by_css_selector("tr:nth-child(1) td+ td")]
                
                #item_name.append(item)
                #item_area.append(area)
                #품종 이름.list_kind_name
                item_kind+=[list_kind_name[ind_item][ind_kind]]*temp_len
                item_name+=[list_item_name[ind_item]]*temp_len
                item_area+=[list_area_name[ind_area]]*temp_len
                count+=1
        #메모리 사용량으로 인해서.
        driver.quit()
        driver=webdriver.Chrome('../driver/chromedriver.exe')
        driver.set_window_size(1024, 600)
    DATA=pd.concat([
        pd.DataFrame(item_name),
        pd.DataFrame(item_kind),
        pd.DataFrame(item_date),
        pd.DataFrame(item_area),
        pd.DataFrame(item_mean),
    ],axis=1)
    DATA.columns= ['품목','품종','날짜','지역','평균가격']
    file_name="".join(['../files/소매가격/',list_item_name[ind_item],'.xlsx'])
    DATA.to_excel(file_name,index=False,)
    print(list_item_name[ind_item],' 출력')
        
print(count,"페이지")



````
