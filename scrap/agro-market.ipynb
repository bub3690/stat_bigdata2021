````python
from selenium import webdriver
#샐래니움 패키지에서 webdriver class를 import한다.
import time
import pandas as pd
import re
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import os
import openpyxl
````

````python
# webdriver 구동하기

# webdriver path지정
driver=webdriver.Chrome('../driver/chromedriver.exe') #webdriver class에서 chrome함수를 불러, 새로운 객체를 또 만들어준다.
# ./ 는 현재 코드가 실행된 현재 위치. 상대경로라고 한다.
````

````python
#url 제어하기

#도매유통 정보시스템으로 이동하자
url="https://at.agromarket.kr/domeinfo/sanTrade.do?pageNo=1&startDateBefore=2019-01-01&endDateBefore=2021-10-03&largeCdBefore=10&midCdBefore=&startDate=2019-01-01&endDate=2021-10-03&whsalCd=110001&cmpCd=&sanCd=&largeCd=10&midCd=01"
driver.get(url)
````

````python
#메인 로직

list_area=[
    "110001",#서울가락  
    "110008",#서울강서  
   "310101" ,#수원  
   "310401" ,#안양  
   "310901" ,#안산  
   "311201" ,#구리  
   "230001" ,#인천남촌  
   "230003" ,#인천삼산  
   "360301" ,#순천  
   "240001" ,#광주각화  
   "240004" ,#광주서부  
   "350402" ,#정읍  
   "350301" ,#익산  
   "350101" ,#전주  
   "250001" ,#대전오정  
   "250003" ,#대전노은  
   "330101" ,#청주  
   "340101" ,#천안  
   "330201" ,#충주  
   "370401" ,#안동  
   "371501" ,#구미  
   "220001" ,#대구북부  
   "380401" ,#진주  
   "380101" ,#창원팔용  
   "380303" ,#창원내서  
   "210001" ,#부산엄궁  
   "210009" ,#부산반여  
   "380201" ,#울산  
   "370101" ,#포항  
   "320201" ,#원주  
   "320101" ,#춘천  
   "320301" ,#강릉  
   "210005" ,#부산국제수산  
    ]#출하지 담는 리스트
list_area_name=[ "서울가락","서울강서","수원","안양","안산","구리","인천남촌" ,"인천삼산","순천",
                "광주각화","광주서부" ,"정읍","익산","전주","대전오정","대전노은","청주" ,"천안","충주","안동","구미",'대구북부'
                ,'진주',"창원팔용","창원내서",'부산엄궁',"부산반여","울산","포항","원주","춘천","강릉","부산국제수산"  ]#출하지 이름담는 리스트
list_cate=["10","10","10","11","12","12","12"]#카테고리 담는 리스트
list_item=["01","13","09","01","03","10","09"]
list_item_name=["배추","갓","미나리","무","쪽파","생강","마늘"]#품목이름을 담는 리스트
count = 0 
for ind_area, area in enumerate(list_area):
    for ind_item,item in enumerate(list_item):
        #1.url 조합. 이동
        url1="https://at.agromarket.kr/domeinfo/sanTrade.do?pageNo=1&startDateBefore=2019-01-01&endDateBefore=2021-10-03&largeCdBefore=10&midCdBefore=&startDate=2019-01-01&endDate=2021-10-03&"
        url2="whsalCd="+area# + 지역
        url3="&cmpCd=&sanCd=&largeCd="+list_cate[ind_item] # + 카테고리
        url4="&midCd="+item# + 품목
        url1=url1+url2+url3+url4

        driver.get(url1)
        try:
            WebDriverWait(driver,timeout=30).until(EC.presence_of_element_located((By.CSS_SELECTOR, "#divMid a+ a")))
        except:
            #데이터가 없는 경우는. 다음으로.
            continue
        time.sleep(5)
        
        #2.excelbox 클릭
        print("다운로드 클릭")
        driver.execute_script("excelDownload();")
        print("완료")
        driver.implicitly_wait(30)
        time.sleep(10)
        file_oldname=os.path.join("C:/Users/dlwhd/Downloads/","출하지별 거래현황 2019-01-01-2021-10-03.xls")
        newname=list_area_name[ind_area]+"-"+list_item_name[ind_item]+".xls"
        file_newname=os.path.join("C:/Users/dlwhd/Downloads/",newname)
        os.rename(file_oldname,file_newname)
        print(area," ",item," ",count,"번 파일")
        count+=1
    
````

````python
#파일들 합치기
for item in list_item_name:
    Temp_data=[]
    for area in list_area_name:
        file_name='../files/전체/'+area+'-'+item+'.xls'
        try:
            Temp_data.append(pd.read_excel(file_name))
        except FileNotFoundError as e:
            #국제수산처럼, 파일이 없는 경우.
            continue
    file_name='../files/도매시장/'+item+'.xlsx'
    pd.concat(Temp_data).to_excel(file_name,index=False)
````
